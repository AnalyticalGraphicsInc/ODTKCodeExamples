# Introduction to Statistic

## Basic Concepts

### Expected Value
In probability theory, the expected value of a random variable is intuitively the long-run average value of repetitions of the experiment it represents. More practically, the expected value of a discrete random variable is the probability-weighted average of all possible values. In other words, each possible value the random variable can assume is multiplied by its probability of occurring, and the resulting products are summed to produce the expected value.

Suppose a random variable $x$ that can take value $x_1$ with probability $p_1$, value $x_2$ with probability $p_2$ etc. The expected value of this random variable is

#### $$ E(x) = x_1p_1+x_2p_2+...+x_nP_n$$

Since all probabilities  $p_i$ add up to one ($p_1+p_2+⋯+p_N=1$), the expected value can be viewed as the weighted average, with $p_i$’s being the weights.
As an example, consider a dice roll. The possible values for x are 1, 2, 3, 4, 5, and 6, each having a probability of 1/6. In this case the expected value is

$$E(x) = 1*\frac{1}{6}+2*\frac{1}{6}+3*\frac{1}{6}+4*\frac{1}{6}+5*\frac{1}{6}+6*\frac{1}{6}= 3.5$$

### Mean
The **mean** is the expected value of an event occurring over all possibilities. It is a quantity representing the 'center' of a collection of numbers, having an intermediate value between the minimum and the maximum value of the collection. Given a population of $n$ values measurements whose value is expressed as $y_{n}$, the mean is defined as: 

#### $$\mu = \frac{1}{n}\sum_{1}^{n} y_{n}$$

For example, consider the following population

$$ \textbf{y} =\left [ 2, 4, 4, 4, 5, 5, 7, 9 \right ]^T$$

The mean is 
$$ \mu = \frac{2+4+4+4+5+5+7+9}{8} = 5$$

----
Which is the difference between the mean and the expected value in practice?
The mean is calculated on data which are already availables, while the expected value is a probabilistic concept that represents the average value expected from a random value considering all the possible outcomes togheter with their probabilities.

For example, consider 10 consecutive dice rolls. Each single roll has an expected value of $3.5$ as seen in the example above: this is the value we *expect* before each roll. Now, imagine that, after the 10 rolls, we got the following outcomes: $7, 4, 11, 9, 7, 6, 2, 8, 3, 5$. In this case the mean is $6.2$, which is different than the expected value. 

---
### Variance
Once the mean is calculated, we can then derive the **variance**, which is a measure of the dispersion around the mean, meaning it is a measure of how far a set of numbers is spread out from their average value.:
  
#### $$\sigma^2 = \frac{1}{N} \sum_{n=1}^{N} \left ( y_{n} - \mu \right )^{2} $$

for the example case we have:

$$ \sigma^2 = \frac{(2-5)^2+(4-5)^2+(4-5)^2+(4-5)^2+(5-5)^2+(5-5)^2+(7-5)^2+(9-5)^2}{8}= \frac{32}{8}=4$$

---
Often the variance is expressed as:

$$\sigma^2 = \frac{1}{N-1} \sum_{n=1}^{N} \left ( y_{n} - \mu \right )^{2} $$

This is called **sample** (or **campionary**) variance. The difference is that in this case the variance is not calculated across the entire population, but just on a 'sample' of it. The practical reason is that, when the population start increasing it becomes difficult to calculate its variance, because we would need a large number of observations. In such a situation, a certain number of observations (a 'sample') are taken and used to describe the entire group

---

### Standard Deviation
Finally, we define the **standard deviation** as the positive square root of the variance:

#### $$ \sigma = +\sqrt{\sigma^2}$$

In the example case we have
$$ \sigma = \sqrt{4}=2$$

### Probability Density Function

The **Probability Density Function**, also named PDF, defines the likelihood that a continuous random variable $x$ will lie within a particular range of values $\eta < x< \eta + d \eta$.

The PDF is expressed as $y=f(x)$; it shall be continous and always > 0.

Suppose that a PDF is defined as
$$ f(x) = 3x^2, 0\leq x\leq 1$$

What the PDF tells us is  that we're likely to have bigger probability that $x$ will occur towards 1 than towards 0. Being $x$ a continous variable, the probability that it will be *exactly* equal to any real number between 0 and 1 is zero, but the PDF will tell us the probability that it will fall between any defined inteval contained in $\left [ 0,1 \right ]$. Defined this interval as $\left [ a,b \right ]$, the probability is evaluated as follow:

$$ \int _{a}^{b} 3x^2 dx = \left [ x^3 \right ]_{a}^{b}$$

Consider the following cases:
- $a=0, b=1$:
$$ \left [ x^3 \right ]_{0}^{1} = 1 - 0 = 1$$
- $a=0, b=0.5$:
$$ \left [ x^3 \right ]_{0}^{0.5} = 0.125 - 0 = 0.125$$
- $a=0.5, b=1$:
$$ \left [ x^3 \right ]_{0.5}^{1} = 1 - 0.125 = 0.875$$

## Normal Distribution
The **normal (or Gaussian) distribution** is a very commonly occurring probability density function. It is a function that tells the probability that any real observation will fall between any two real limits or real numbers, as the curve approaches zero on either side. 
The normal distribution is extremely useful because of the **central limit theorem**, which states that a large number of individual estimates having independent errors will converge to a normal (Gaussian) distribution.

The general formula for the PDF of the Gaussian distribution is:



#### $$ f(x) =  \frac{1}{\sigma \sqrt{2 \pi}}exp \left[- \frac{1}{2} \left ( \frac{x - \mu}{\sigma} \right )^{2} \right]$$

, so the normal distribution is only function of the mean $\mu$ and the standard deviation $\sigma$.

```python
from statistic import *
from ipywidgets import interact, FloatSlider
%matplotlib inline

def plt_g(mu,sigma):
    plt.figure(figsize=(12,4))
    xs = np.arange(2, 8, 0.01)
    plt.plot(xs,  norm.pdf(xs, mu, sigma), label='Normal Distribution', linewidth=3)
    plt.ylim(0, 1.2)
    plt.grid(True)
    plt.legend()
    plt.show()

interact(plt_g, mu=FloatSlider(value=5, min=3, max=7),sigma=FloatSlider(value = .6, min=.4, max=1.));
```

Like any other PDF curves, the area below the curve tells us the probability that a specific event (e.g., the value of a measure) will occur between specific limits.While the probability is 100% when the limits are $(+ \infty, - \infty)$, we express the probablity in terms of multiplying factors of $\sigma$:

- 1-$\sigma$: 68.27%
- 2-$\sigma$: 95.45%
- 3-$\sigma$: 99.73%

```python
from statistic import *
%matplotlib widget

########### input data ###############
mean = 18    # population mean
sigma = 1   # standard deviation
nSigma = 2   # interval to consider
######################################

PlotNormalDistribution(mean, sigma, nSigma)
```

## Data Fusion

A key property of the Gaussian function is exploited at this point: the product of two gaussian functions is another gussian function. 

If we define $\mu_1$, $\sigma_1$ the parameters of the first distribution and $\mu_2$, $\sigma_2$ the parameters of the second distribution, multiplying the two will get:

$$ \mu_m = \frac{\mu_1 \sigma_{2}^{2} + \mu_2 \sigma_{2}^{2}}{\sigma_{1}^{2} + \sigma_{1}^{2}}$$

and

$$ \sigma_m =  \sqrt{\frac{\sigma_{1}^{2} \sigma_{2}^{2}}{\sigma_{1}^{2} + \sigma_{2}^{2}}}$$

This property is used to improve the quality of the measurements by using two different sensors to measure the same event; this approach is also called **data fusion**.

 ## Covariance and Covariance Matrix
As stated above, the variance is an indicator of the variation of a population of random value around its mean. In a similar fashion, the **covariance** is a measure of how much different populations of random variables vary relevant to each other.

Defining  $ \textbf{y}_1 =\left [ y_{11}, y_{12}, ..., y_{1m }\right ]^T,  \textbf{y}_2 =\left [ y_{21}, y_{22}, ... , y_{2m }\right ]^T, ... , \textbf{y}_n =\left [ y_{n1}, y_{n2}, ..., y_{nm }\right ]^T$ $n$ datasets of random variables each containing $m$ samples, we define the variance on the $i^{th}$ population as:

#### $$\sigma_i^2 = \frac{1}{m-1} \sum_{k=1}^{m} \left ( y_{ik} - \mu_i \right )^{2}\quad \quad  i=1...n$$

The covariance $\sigma_{ij}$ relates the $i^{th}$ and $j^{th}$ dataset, highligting if there is a correlation between the two:

#### $$\sigma_{ij} = \frac{1}{m-1} \sum_{k=1}^{m} \left ( y_{ik} - \mu_i \right ) \left ( y_{jk} - \mu_j\right )=\rho_{ij} \sigma_i \sigma_j \quad \quad  i,j=1...n$$

, where $\mu_i$, $\mu_j$ are the means of the $i^{th}$ and $j^{th}$ dataset and $\rho_{ij}$ is the **correlation coefficient**. By definition, it takes a value between -1 and 1 and is a measure of the dependance of the two datasets of random variables. 

Calculating all the covariances we get an $n$x$n$ symmetrix matrix called **covariance matrix**:

#### $$ \mathbf{C}=\begin{bmatrix}
\sigma_1^2 & \sigma_{12} &  ...  & \sigma_{1n}\\
\sigma_{21} & \sigma_2^2 &  ...  & \sigma_{1n}\\ &  &  &\\
... & ... & ... &...\\
\sigma_{1n} & \sigma_{2n} & ... & \sigma_n^2\\
\end{bmatrix}, \quad \sigma_{ij}=\sigma_{ji}$$

### Example 1
Consider the following table, where $n=2$ and $m=5$:

| Student | X - Effort spent (hr) | Y - Score |
|---------|-----------------------|-----------|
| 1       | 10                    | 70        |
| 2       | 15                    | 80        |
| 3       | 20                    | 90        |
| 4       | 25                    | 95        |
| 5       | 30                    | 100       |


The $X$ dataset represents the number of hours a student spent to study a specific subject, and $Y$ is the score he got for that subject. The means are:

$$ \mu_X = \frac{10+15+20+25+30}{5} = 20, \quad \quad \mu_Y = \frac{70+80+90+95+100}{5} = 87$$

The variances are:

$$ \sigma_X^2 = \frac{1}{5-1}[(10-20)^2 +(15-20)^2 +(20-20)^2 +(25-20)^2 +(30-20)^2 ] = 62.5$$
$$ \sigma_Y^2 = \frac{1}{5-1}[(70-87)^2 +(80-87)^2 +(90-87)^2 +(95-87)^2 +(100-87)^2 ] = 145$$

The covariances are:

$$ \sigma_{XY}=\sigma_{YX}= \frac{1}{5-1}[(10-20)(70-87)+(15-20)(80-87)+(20-20)(90-87)+(25-20)(95-87)+(30-20)(100-87)]=93.75$$

...and the covariance matrix is

$$ \mathbf{C}= \begin{bmatrix}
 62.5&  93.75\\
93.75 &  145\\
\end{bmatrix}$$

Finally, the correlation coefficient is

$$\rho_{XY}=\frac{\sigma_{XY}}{\sigma_{X} \sigma_{Y}}=\frac{93.75}{\sqrt{62.5} \sqrt{145}}=0.985$$

We see how the two dataset are highly correlated (correlation coefficient close to 1) with positive sign, meaning that for increasing number of studying hours we have increasing scores.

## Autocorrelation

**Autocorrelation**, also known as serial correlation, is a statistical concept that refers to the correlation of a signal with a delayed copy of itself as a function of delay. In a time series context, autocorrelation can be thought of as the correlation between a series and its lagged values. If a series is autocorrelated, it means that the values in the series are not independent of each other.

Mathematically, the **autocorrelation function** (ACF) is defined as the correlation between the elements of a series and others from the same series separated from them by a given interval. We can write this for real-valued discrete signals as

#### $$R_{ff}(l)= \sum_{n=0}^{N}f(n)f(n-l)$$

The degree of correlation is normalized in the $\left [ -1, 1 \right]$ interval:

- A correlation coefficient close to 1 indicates a strong positive autocorrelation. That is, a high value in the time series is likely to be followed by another high value, and a low value is likely to be followed by another low value
- A correlation coefficient close to -1 indicates a strong negative autocorrelation. That is, a high value in the time series is likely to be followed by a low value, and vice versa
- A correlation coefficient closer to 0 indicates no correlation. That is, the values in the time series appear to be random and do not follow a discernible pattern

The above plot shows the autocorrelation function of two datasets, one composed by random variables and the other defined by a $sin(\omega t)$ relationship
